## Plotting sine and cosine waves

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

Working with shapes in a plot:

```python
plt.plot(X, C, 'r--')
plt.plot(X, S, 'g--')
plt.show()

# A few examples:
# To get squares: '<color-initial>s'
# To get triangles: '<color-initial>^'
```

Controlling the minute parameters related to the plots using the `setup()` command:

```python
lines = plt.plot(X, S)
# use keyword args
plt.setp(lines, color='r', linewidth=2.0)
plt.show()
```

Using labels, titles and annotations:

```python
lines = plt.plot(X, S)
# use keyword args
plt.setp(lines, color='r', linewidth=2.0)
plt.xlabel('x')
plt.ylabel('y')
plt.title('It is a Sine wave!')

# Note: Annotation is not essential majority of the times,
# there's always a trial&error invovled in getting it right
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.show()
```

## A Bar Chart

```python
y = [3, 10, 7, 5, 3, 4.5, 6, 8.1]
N = len(y)
x = range(N)
width = 1/1.5
plt.bar(x, y, width, color="blue")
save_fig("bar_chart")
```

## A Stacked Bar Chart

```python
n = 5
data1 = np.array(range(n)) + np.random.rand(n)
data2 = np.random.rand(n) + 1

bar_locations = np.arange(n)
plt.bar(bar_locations, data1)
plt.bar(bar_locations, data2, bottom=data1, color='r')
plt.title("Stacked Bar Charts")
save_fig("stacked_bar_chart")
plt.show()
```

## A Grouped Bar Chart

```python
n = 5
data1 = np.array(range(n)) + np.random.rand(n)
data2 = (np.array(range(n)) + np.random.rand(n)) / 2

bar_width = 0.4  # default: 0.8
bar_locations = np.arange(n)

plt.bar(bar_locations, data1, bar_width)
plt.bar(
	bar_locations - bar_width, 
	data2, 
	bar_width, 
	color='r'
)

plt.title("Grouped Bar Charts")
save_fig("grouped_bar_chart")
plt.show()
```

## Q-Q plots

```python
import numpy as np
import numpy.random as random

p = random.poisson(lam=10, size=4000)
m = np.mean(p)
s = np.std(p)
n = random.normal(loc=m, scale=s, size=p.shape)

a = m-4*s
b = m+4*s

plt.figure()
plt.plot(np.sort(n), np.sort(p), 'o', color='red')
plt.plot([a,b], [a,b], 'k-')
plt.xlim(a,b)
plt.ylim(a,b)
plt.xlabel('Normal Distribution')
plt.ylabel('Poisson Distribution with $\lambda=10$')
plt.grid(True)
save_fig("q_q_plot")
plt.show()
```
